```python
funcion = sin(x) #Funcion para serie de Taylor
funcion
```

```python
a = 0 #Alrededor de 0
n = 4+1 #A orden 4
```

```python
funcion.series(x, a, n ).removeO() #funcion.series(variable, alrededor , orden)
```

```python
def taylor(funcion, a, n ):
  return funcion.series(x,a, n).removeO()
```

```python
funcion = sin(x)
a = 0
n = 4+1
taylor(funcion, a, n)
```

```python
#Crear una función para graficar 3 aproximaciones de Taylor
# a = punto alrededor de la aproximación
#n1 = orden aproximación 1
#n2 = orden aproximación 2
#n3 = orden aproximación 3

def graficar_taylor(funcion, a, n1, n2, n3, n4,n5, inicio, final):
  p = plot(funcion, taylor(funcion, a, n1), taylor(funcion, a, n2),
           taylor(funcion, a, n3),taylor(funcion, a, n4),
           taylor(funcion, a, n5), (x, inicio, final), legend = True, show=False )
  
  p[0].line_color = "blue" # Color funcion original
  p[1].line_color = "green" #Color aproximación 1
  p[2].line_color = "firebrick" #Color aprox 2
  p[3].line_color = "black" #Color aprox 3
  p[4].line_color = "brown" #Color aprox 4
  p[5].line_color = "yellow" #Color aprox 5

  p.title = "Serie de Taylor" #Titulo de la grafica
  p.show() # Mostrar grafica
```

```python
funcion = sin(x)
a = 0
inicio = -2
fin = 2
n1 = 2+1
n2 = 4+1
n3 = 6+1
graficar_taylor(funcion, a, n1, n2, n3, inicio, final)
```
